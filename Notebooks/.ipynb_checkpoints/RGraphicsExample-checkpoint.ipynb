```fsharp
open RDotNet
open RProvider
open RProvider.``base``
open RProvider.graphics

let widgets = [3;8;12;15;19;18;18;20;]

/// helper function to evaluate a string in R
let eval (text:string) = R.eval(R.parse(namedParams ["text", text ]))

eval ("""win.metafile(filename="",pointsize = 12, width = 6, height = 6)""") |> ignore
R.plot(widgets)
eval "dev.off()" |> ignore
Util.Image(ClipboardHelper.getPng(1000,1000)) |> Display
```

```fsharp
open RDotNet
open RProvider
open RProvider.``base``
open RProvider.graphics

/// helper function to evaluate a string in R
let eval (text:string) = R.eval(R.parse(namedParams ["text", text ]))

let plot (name:string) (chart : unit -> RDotNet.SymbolicExpression) =
    eval <| System.String.Format("""jpeg("{0}.jpg")""", name) |> ignore
    chart()  |> ignore
    eval "dev.off()"  |> ignore
    Util.Image (name + ".jpg") |> Display
```
